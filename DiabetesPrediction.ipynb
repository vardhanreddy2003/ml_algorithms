```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=["Outcome"]),df["Outcome"],test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train_transformed=StandardScaler().fit_transform(X_train)
X_test_transformed=StandardScaler().fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred=rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```
