```python
import numpy as np 
import pandas as pd
```

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
X,y=make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20,random_state=0)
```

```python
y.shape
```

```python
sns.scatterplot(x=X[:,0],y=y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

```python
lr=LinearRegression()
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
rr1=Ridge(alpha=10)
rr1.fit(X,y)
print(rr1.coef_)
print(rr1.intercept_)
```

```python
rr2=Ridge(alpha=100)
rr2.fit(X,y)
print(rr2.coef_)
print(rr2.intercept_)
```

```python

sns.scatterplot(x=X[:,0],y=y,color="blue")
sns.lineplot(x=X[:,0],y=lr.predict(X),label="lr",color="yellow")
sns.lineplot(x=X[:,0],y=rr1.predict(X),label="rr1",color="green")
sns.lineplot(x=X[:,0],y=rr2.predict(X),label="rr2",color="red")

plt.show()
```
